This notebook demonstrates dataset exploration and model inference.

```python
from src.data import make_dataloaders
from src.predict import load_model, predict_image
from src.model import get_device
import matplotlib.pyplot as plt
from pathlib import Path

DATA_DIR = 'data/plant_village'
train_loader, val_loader, _, classes = make_dataloaders(DATA_DIR, batch_size=8)
classes
```

```python
# Show sample images
images, labels = next(iter(train_loader))
plt.figure(figsize=(10, 6))
for i in range(6):
    plt.subplot(2, 3, i+1)
    img = images[i].permute(1,2,0).numpy()
    plt.imshow(img)
    plt.title(classes[labels[i].item()])
    plt.axis('off')
plt.show()
```

```python
# Load trained model
checkpoint = 'checkpoints/best_model.pth'
model, device = load_model(checkpoint, classes)
model
```

```python
# Predict on a random image
sample_img = list(Path(DATA_DIR).rglob('*.jpg'))[0]
pred, conf = predict_image(str(sample_img), model, classes, device)
pred, conf
```
